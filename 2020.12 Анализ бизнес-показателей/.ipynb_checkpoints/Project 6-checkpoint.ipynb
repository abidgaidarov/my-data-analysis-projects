## –®–∞–≥ 2. –ü–æ—Å—Ç—Ä–æ–∏–º –æ—Ç—á—ë—Ç—ã –∏ –ø–æ—Å—á–∏—Ç–∞–µ–º –º–µ—Ç—Ä–∏–∫–∏

### 2.1. –ü—Ä–æ–¥—É–∫—Ç.

#### 2.1.1. –û–ø—Ä–µ–¥–µ–ª–∏–º —Å–∫–æ–ª—å–∫–æ –ª—é–¥–µ–π –ø–æ–ª—å–∑—É—é—Ç—Å—è –≤ –¥–µ–Ω—å (DAU), –Ω–µ–¥–µ–ª—é (WAU), –º–µ—Å—è—Ü (MAU), –∞ —Ç–∞–∫–∂–µ —Ä–µ–≥—É–ª—è—Ä–Ω–æ—Å—Ç—å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å–µ—Ä–≤–∏—Å–∞ (sticky factor).

–î–æ–±–∞–≤–∏–º –≤ —Ç–∞–±–ª–∏—Ü—É visits —Å—Ç–æ–ª–±—ã —Å –¥–Ω–µ–º/–Ω–µ–¥–µ–ª–µ–π/–º–µ—Å—è—Ü–µ–º —Å–µ—Å—Å–∏–∏.

```python
visits['date'] = visits['start_ts'].dt.date
visits['week'] = visits['start_ts'].dt.strftime('%Y-%W')
visits['month'] = visits['start_ts'].dt.strftime('%Y-%m')
```

–û–ø—Ä–µ–¥–µ–ª–∏–º DAU, WAU, MAU:

```python
dau = visits.pivot_table(index='date', values='user_id', aggfunc='nunique')['user_id'].mean()
wau = visits.pivot_table(index='week', values='user_id', aggfunc='nunique')['user_id'].mean()
mau = visits.pivot_table(index='month', values='user_id', aggfunc='nunique')['user_id'].mean()
print('DAU —Ä–∞–≤–Ω–æ {:.2f}.'.format(dau))
print('WAU —Ä–∞–≤–Ω–æ {:.2f}.'.format(wau))
print('MAU —Ä–∞–≤–Ω–æ {:.2f}.'.format(mau))
```

–û–ø—Ä–µ–¥–µ–ª–∏–º sticky factor.

```python
sticky_factor = dau/mau*100
print('Sticky factor —Ä–∞–≤–µ–Ω {:.2f}.'.format(sticky_factor))
```

#### 2.1.2. –û–ø—Ä–µ–¥–µ–ª–∏–º —Å–∫–æ–ª—å–∫–æ —Å–µ—Å—Å–∏–π –≤ –¥–µ–Ω—å.

```python
sessions_amount = visits.pivot_table(index='date', values='start_ts', aggfunc='count')['start_ts'].mean()
print('–°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ —Å–µ—Å—Å–∏–π –≤ –¥–µ–Ω—å —Ä–∞–≤–Ω–æ {:.2f}.'.format(sessions_amount))
sessions_per_user = sessions_amount/dau
print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–µ—Å—Å–∏–π –≤ –¥–µ–Ω—å –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è {:.2f}.'.format(sessions_per_user))
```

#### 2.1.3. –û–ø—Ä–µ–¥–µ–ª–∏–º —Å–∫–æ–ª—å–∫–æ –¥–ª–∏—Ç—Å—è –æ–¥–Ω–∞ —Å–µ—Å—Å–∏—è (ASL).

–î–æ–±–∞–≤–∏–º –≤ —Ç–∞–±–ª–∏—Ü—É visits —Å—Ç–æ–ª–±–µ—Ü —Å –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é —Å–µ—Å—Å–∏–∏ –≤ —Å–µ–∫—É–Ω–¥–∞—Ö.

```python
visits['duration'] = (visits['end_ts'] - visits['start_ts']).dt.seconds
```

<div style="border:solid orange 2px; padding: 20px"> <b>–ß—Ç–æ –º–æ–∂–Ω–æ —É–ª—É—á—à–∏—Ç—å</b><br>
–ò—Å–ø–æ–ª—å–∑—É–µ–º—ã–π —Ç–æ–±–æ–π –º–µ—Ç–æ–¥ <code>seconds</code> –æ—Ç–ª–∏—á–Ω–æ —Ä–∞–±–æ—Ç–∞–µ—Ç, –µ—Å–ª–∏ –µ—Å—Ç—å —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å, —á—Ç–æ –¥–∞—Ç—ã –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ (–¥–∞—Ç–∞ –æ–∫–æ–Ω—á–∞–Ω–∏—è –±–æ–ª—å—à–µ –¥–∞—Ç—ã –Ω–∞—á–∞–ª–∞), –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –º–µ—Ç–æ–¥ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ–ª-–≤–æ —Å–µ–∫—É–Ω–¥ –≤ –æ–¥–Ω–æ–º –¥–Ω–µ.

–ü–æ—Å–º–æ—Ç—Ä–∏ –Ω–∞ –º–µ—Ç–æ–¥ `total_seconds()` - –æ–Ω –≤ –ª—é–±–æ–º —Å–ª—É—á–∞–µ–º –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ–ª-–≤–æ —Å–µ–∫—É–Ω–¥. –ü–æ–ø—Ä–æ–±—É–π —Å—Ä–∞–≤–Ω–∏—Ç—å –æ–±–∞ –º–µ—Ç–æ–¥–∞, –¥—É–º–∞—é, –Ω–∞–π–¥–µ—Ç—Å—è –∏–Ω—Ç–µ—Ä–µ—Å–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞–Ω–Ω—ã—Ö.</div> <br>

```python
visits['duration2'] = (visits['end_ts'] - visits['start_ts']).dt.total_seconds()
```

<div style="border:solid green 2px; padding: 20px"> <b>–û—Ç–ª–∏—á–Ω–∞—è —Ä–∞–±–æ—Ç–∞ 2</b><br>üëç</div> 

–ü—Ä–æ–≤–µ—Ä–∏–º —Ö–∞—Ä–∞–∫—Ç–µ—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã.

```python
plt.figure(figsize=(15,4))
plt.xticks(np.arange(0, 2000, step=100))

visits_desktop = visits[visits['device'] == 'desktop']['duration']
visits_desktop.hist(range=(0,2000), bins=100, label='desktop')

visits_touch = visits[visits['device'] == 'touch']['duration']
visits_touch.hist(range=(0,2000), bins=100, label='touch')

plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Å–µ—Å—Å–∏–∏', fontsize = 14, weight = 'bold')
plt.xlabel('–ü—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å, —Å–µ–∫')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.legend()
```

```python
print('ASL desktop –≤–µ—Ä—Å–∏–∏ —Ä–∞–≤–µ–Ω {:.2f}.'.format(int(visits_desktop.mode())))
print('ASL touch –≤–µ—Ä—Å–∏–∏ —Ä–∞–≤–µ–Ω {:.2f}.'.format(int(visits_touch.mode())))
```

#### 2.1.4. –û–ø—Ä–µ–¥–µ–ª–∏–º –∫–∞–∫ —á–∞—Å—Ç–æ –ª—é–¥–∏ –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç—Å—è (retention rate).

–û–ø—Ä–µ–¥–µ–ª–∏–ª–∏, —á—Ç–æ 77% –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∏—Å—å —Å–µ—Ä–≤–∏—Å–æ–º –ª–∏—à—å –æ–¥–Ω–∞–∂–¥—ã.

```python
first_session = visits.groupby(['user_id'])['date'].min()
first_session.name = 'first_session'
visits = visits.join(first_session, on='user_id', how='left')
```

```python
visits['first_session_month'] = visits['first_session'].astype('datetime64[M]')
visits['session_month'] = visits['date'].astype('datetime64[M]')
```

```python
visits['lifetime'] = visits['session_month'] - visits['first_session_month']
visits['lifetime'] = visits['lifetime'] / np.timedelta64(1,'M')
visits['lifetime'] = visits['lifetime'].round().astype('int')
```

```python
cohorts = (visits
.groupby(['first_session_month','lifetime'])
.agg({'user_id':'nunique'}).reset_index()
)
```

```python
cohort_users = cohorts[cohorts['lifetime'] == 0][['first_session_month','user_id']]
cohort_users = cohort_users.rename(columns = {'user_id': 'cohort_users'})
cohorts = cohorts.merge(cohort_users, on = 'first_session_month')
cohorts['retention'] = cohorts['user_id'] / cohorts['cohort_users']
```

```python
cohorts['first_session_month_str'] = cohorts['first_session_month'].dt.strftime('%Y-%m-%d')
```

```python
ret_rate = cohorts.query('lifetime != 0').pivot_table(index = 'first_session_month_str',
columns = 'lifetime',
values = 'retention',
aggfunc = 'sum')
```

```python
sns.set(style = 'white')
plt.figure(figsize = (15, 4))
plt.title('Retention Rate', fontsize = 14, weight = 'bold')
fig = sns.heatmap(ret_rate,
annot = True, fmt = '.1%',
linewidths = 1, linecolor = 'white', cmap = 'rocket')
plt.xlabel('–í–æ–∑—Ä–∞—Å—Ç')
plt.ylabel('–ö–æ–≥–æ—Ä—Ç–∞')
plt.show(fig)
```

–†–∞—Å—á–∏—Ç–∞–µ–º –≤—Ä–µ–º—è –ø–µ—Ä–≤–æ–π —Å–µ—Å—Å–∏–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.

#### 2.2.4. –û–ø—Ä–µ–¥–µ–ª–∏–º —Å–∫–æ–ª—å–∫–æ –¥–µ–Ω–µ–≥ –ø—Ä–∏–Ω–æ—Å—è—Ç (LTV).

LTV (–æ—Ç –∞–Ω–≥–ª. lifetime value) ‚Äî ¬´–ø–æ–∂–∏–∑–Ω–µ–Ω–Ω–∞—è¬ª —Ü–µ–Ω–Ω–æ—Å—Ç—å –∫–ª–∏–µ–Ω—Ç–∞. –≠—Ç–æ ‚Äî –∏—Ç–æ–≥–æ–≤–∞—è —Å—É–º–º–∞ –¥–µ–Ω–µ–≥, –∫–æ—Ç–æ—Ä—É—é –æ–¥–∏–Ω –∫–ª–∏–µ–Ω—Ç –ø—Ä–∏–Ω–æ—Å–∏—Ç –≤ —Å—Ä–µ–¥–Ω–µ–º —Å–æ –≤—Å–µ—Ö —Å–≤–æ–∏—Ö –ø–æ–∫—É–ø–æ–∫. –û–ø—Ä–µ–¥–µ–ª–∏–º –µ–µ.

```python
orders['first_order_month'] = orders['first_order_time'].astype('datetime64[M]')
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')
```

–ü–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–æ–≤—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π (n_buyers) –∑–∞ –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü:

```python
cohort_sizes = orders.groupby('first_order_month').agg({'user_id': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']
```

–°–≥—Ä—É–ø–ø–∏—Ä—É–µ–º —Ç–∞–±–ª–∏—Ü—É –∑–∞–∫–∞–∑–æ–≤ –ø–æ –º–µ—Å—è—Ü—É –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏ –∏ –º–µ—Å—è—Ü—É –∫–∞–∂–¥–æ–≥–æ –∑–∞–∫–∞–∑–∞ –∏ —Å–ª–æ–∂–∏–º –≤—ã—Ä—É—á–∫—É.

```python
cohorts = orders.groupby(['first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()
```

–î–æ–±–∞–≤–∏–º –≤ —Ç–∞–±–ª–∏—Ü—É cohorts –¥–∞–Ω–Ω—ã–µ –æ —Ç–æ–º, —Å–∫–æ–ª—å–∫–æ –ª—é–¥–µ–π –ø–µ—Ä–≤—ã–π —Ä–∞–∑ —Å–æ–≤–µ—Ä—à–∏–ª–∏ –ø–æ–∫—É–ø–∫—É –≤ –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü:

```python
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
```

–ù–∞–π–¥–µ–º –≤–∞–ª–æ–≤—É—é –ø—Ä–∏–±—ã–ª—å, —É–º–Ω–æ–∂–∏–≤ –≤—ã—Ä—É—á–∫—É –Ω–∞ –º–∞—Ä–∂–∏–Ω–∞–ª—å–Ω–æ—Å—Ç—å. –¢–∞–∫ –∫–∞–∫ LTV ‚Äî –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å, —Ä–∞—Å—á–∏—Ç–∞–µ–º –µ–≥–æ –ø–æ ¬´–≤–æ–∑—Ä–∞—Å—Ç–Ω—ã–º¬ª –∫–æ–≥–æ—Ä—Ç–∞–º, –≥–¥–µ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö ‚Äî –Ω–µ –º–µ—Å—è—Ü –∑–∞–∫–∞–∑–∞, –∞ –≤–æ–∑—Ä–∞—Å—Ç (age) –∫–æ–≥–æ—Ä—Ç—ã.

```python
margin_rate = 1
    
report['gp'] = report['revenue'] * margin_rate
report['age'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
```

–ù–∞–∫–æ–Ω–µ—Ü, –Ω–∞–π–¥—ë–º LTV. –†–∞–∑–¥–µ–ª–∏–º –≤–∞–ª–æ–≤—É—é –ø—Ä–∏–±—ã–ª—å –∫–æ–≥–æ—Ä—Ç –∑–∞ –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü –Ω–∞ –æ–±—â–µ–µ —á–∏—Å–ª–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –∫–∞–∂–¥–æ–π –∫–æ–≥–æ—Ä—Ç–µ. –î–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–æ–≥–æ—Ä—Ç—ã –≤ –≤–∏–¥–µ —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã:

```python
report['ltv'] = report['gp'] / report['n_buyers']
    
ltv = report.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='ltv', 
        aggfunc='mean').cumsum(axis=1)
```

–ü–æ—Å—Ç—Ä–æ–∏–º –¥–∏–∞–≥—Ä–∞–º–º—É.

```python
sns.set(style = 'white')
plt.figure(figsize = (15, 4))
plt.title('LTV', fontsize = 14, weight = 'bold')
fig = sns.heatmap(ltv,
annot = True, fmt = '.2f',
linewidths = 1, linecolor = 'white', cmap = 'rocket')
plt.xlabel('–í–æ–∑—Ä–∞—Å—Ç')
plt.ylabel('–ö–æ–≥–æ—Ä—Ç–∞')
plt.show(fig)
```

–í —Å–≤—è–∑–∏ —Å —Ç–µ–º, —á—Ç–æ 77% –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∑–∞—Ö–æ–¥—è—Ç –Ω–∞ —Å–µ—Ä–≤–∏—Å –ª–∏—à—å –æ–¥–Ω–∞–∂–¥—ã, —Å–∞–º—ã–µ –±–æ–ª—å—à–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏—Ä–æ—Å—Ç–∞ LTV –≤ –∫–∞–∂–¥–æ–π –∏–∑ –∫–æ–≥–æ—Ä—Ç - –≤ –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü. LTV –ø–µ—Ä–≤–æ–π –∫–æ–≥–æ—Ä—Ç—ã - 11.88 —É.–µ.

<div style="border:solid orange 2px; padding: 20px"> <b>–ß—Ç–æ –º–æ–∂–Ω–æ —É–ª—É—á—à–∏—Ç—å</b><br>
–î–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –∫–æ–≥–æ—Ä—Ç –º–æ–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å –Ω–µ–∫–∏–π –≥–æ—Ä–∏–∑–æ–Ω—Ç, –Ω–∞–ø—Ä–∏–º–µ—Ä 3\6\9 –º–µ—Å—è—Ü–µ–≤ –∏ —Å–¥–µ–ª–∞—Ç—å —Å—Ä–µ–∑, –¥–ª—è —Ç–µ—Ö –∫–æ–≥–æ—Ä—Ç –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–æ–∂–∏–ª–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ –∫–æ–ª-–≤–æ –≤—Ä–µ–º–µ–Ω–∏</div> <br>

```python
costs['dt_month'] = costs['dt'].astype('datetime64[M]')
costs_pivot3 = costs.pivot_table(columns='dt_month', values='costs', index='source_id', aggfunc='sum')

def costs_plot(source_id):
    data = costs_pivot3.loc[source_id]

    plt.plot(data.index, data.values)
    
plt.figure(figsize = (15, 4))

plt.grid(True)

source_ids = [1, 2, 3, 4, 5, 9, 10]

for id in source_ids:
    costs_plot(id)
    
plt.legend(['1', '2', '3', '4', '5', '9', '10'])

plt.title('–ì—Ä–∞—Ñ–∏–∫ –∑–∞—Ç—Ä–∞—Ç –Ω–∞ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤ –ø–æ –º–µ—Å—è—Ü–∞–º', fontsize = 14, weight = 'bold')
plt.xlabel('–ú–µ—Å—è—Ü—ã')
plt.ylabel('–ó–∞—Ç—Ä–∞—Ç—ã, —É.–µ.')
```

#### 2.3.2. –û–ø—Ä–µ–¥–µ–ª–∏–º —Å–∫–æ–ª—å–∫–æ —Å—Ç–æ–∏–ª–æ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏–µ –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –∏–∑ –∫–∞–∂–¥–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞.

–û–±—ä–µ–¥–µ–Ω–∏–º —Ç–∞–±–ª–∏—Ü—ã visits –∏ orders –ø–æ user_id, —Å —Ü–µ–ª—å—é –ø—Ä–∏—Å–≤–æ–µ–Ω–∏—è –∫–∞–∂–¥–æ–º—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é –∏—Å—Ç–æ—á–Ω–∏–∫–∞, –æ—Ç–∫—É–¥–∞ –æ–Ω –≤–ø–µ—Ä–≤—ã–µ –ø–µ—Ä–µ—à–µ–ª –Ω–∞ —Å–∞–π—Ç.

```python
combined = pd.merge(orders, visits, left_on='user_id', right_on='user_id')[['first_order_month','source_id', 'user_id']]
```

```python
cohorts_source = combined.pivot_table(index='first_order_month', columns='source_id', values='user_id', aggfunc='nunique')
```

```python
costs_pivot4 = costs.pivot_table(index='dt_month', columns='source_id', values='costs', aggfunc='sum')
```

–†–∞—Å—Å—á–∏—Ç–∞–µ–º CAC –∫–∞–∂–¥–æ–π –∫–æ–≥–æ—Ä—Ç—ã –ø–æ –∫–∞–∂–¥–æ–º—É –∏—Å—Ç–æ—á–Ω–∏–∫—É.

```python
cac = costs_pivot4/cohorts_source

sns.set(style = 'white')
plt.figure(figsize = (15, 4))
plt.title('CAC', fontsize = 14, weight = 'bold')
fig = sns.heatmap(cac,
annot = True, fmt = '.2f',
linewidths = 1, linecolor = 'white', cmap = 'rocket')
plt.xlabel('–ò–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä —Ä–µ–∫–ª–∞–º–Ω–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞')
plt.ylabel('–ö–æ–≥–æ—Ä—Ç–∞')
plt.show(fig)
```

–†–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π CAC –∫–∞–∂–¥–æ–π –∫–æ–≥–æ—Ä—Ç—ã –±–µ–∑ —É—á–µ—Ç–∞ –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤.

```python
cac_mean = cac.mean(axis=1) 
cac_mean.name = 'cac_mean'
cac_mean
```

```python
report['romi'] = report['ltv']/report['cac_mean']
```

```python
romi = report.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='romi', 
        aggfunc='mean').cumsum(axis=1)
```
